```python
#SÖZLÜKLER
#Sözlükler, anahtar-değer çiftlerini depolayan veri yapılarıdır. Her anahtar benzersizdir ve bir değere karşılık gelir. 
#Sözlükler, süslü parantez {} içinde tanımlanır ve anahtar-değer çiftleri virgülle ayrılır. 
# Sözlükler sırasızdır, değiştirilebilir ve tekrar eden anahtarlara izin vermezler. 
# Anahtarlar genellikle string türünde olurken, değerler herhangi bir veri türünde olabilir.
#key - value pairing
```

```python
fruit_dictionary = {"banana":100,"apple":150}

fruit_dictionary["apple"]
fruit_dictionary["banana"]
```

```python
fruit_dictionary = {"banana":100,"apple":150}

print(fruit_dictionary.keys()) #sözlüğün anahtarlarını verir
print(fruit_dictionary.values()) #sözlüğün değerlerini verir
print(fruit_dictionary.items()) #sözlüğün anahtar-değer çiftlerini verir
```

```python
fruit_dictionary = {"banana":100,"apple":150}

list(fruit_dictionary.keys())
```

```python
fruit_dictionary = {"banana":100,"apple":150}

fruit_dictionary["banana"] = 120 #değeri güncelleme
print(fruit_dictionary)
fruit_dictionary["orange"] = 80 #yeni anahtar-değer çifti ekleme
print(fruit_dictionary)
```

```python
fruit_dictionary = {"banana":100,"apple":150}

fruit_dictionary.get("apple",0) #anahtarın değerini döndürür, anahtar yoksa 0 döndürür
fruit_dictionary.get("grape",0) #anahtar yok, 0 döndürür
```

```python
my_mixed_dictionary = {"key1":100,"key2":3.14,"key3":[10,20,30]}

my_mixed_dictionary["key3"][1]
```

```python
last_dictionary = {"k1":10,"k2":[10,20,30,40,50],"k3":"string","k4":{"a":100,"b":200}}

print(last_dictionary["k2"][2])
print(last_dictionary["k4"]["b"])
```
