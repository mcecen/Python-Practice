```python
#DEMETLER(Tuples)
# Demetler, listeler gibi sıralı veri yapılarıdır ancak demetler değiştirilemez (immutable) ve parantez () kullanılarak tanımlanır. 
# Demetler, birden fazla değeri tek bir değişkende saklamak için kullanılır ve genellikle farklı türde verileri içerebilirler. 
# Demetler, listelere göre daha az bellek kullanır ve daha hızlıdır, bu nedenle değiştirilmeyecek veriler için tercih edilirler.
```

```python
myTuple = (10,"a","b",3.14)

print(myTuple[0])
print(myTuple.index("a"))
print(myTuple.count("a"))
```

```python
resultTuple = (10,100,20)

resultList = list(resultTuple)
print(resultList)
```
