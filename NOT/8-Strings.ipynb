```python
name="Mustafa"
print(name) 
```

```python
name="cecen mustafa cecen"
name[0]
name[-1]
```

```python
name="cecen mustafa cecen"
#[starting index:stopping index:stepping size]
name[2::]
name[::-1]
name[1:2:3]
```

```python
#capitalize(): bir karakter dizisinin ilk karakterini büyük harfe çevirir.
name="Mustafa"
name.capitalize()
print(name)
```

```python
#count(): bir karakter dizisinde belirli bir karakterin kaç kez geçtiğini sayar.
name="Mustafa"
name.count("a")
```

```python
#replace(): bir karakter dizisi içindeki karakterleri başka karakterle değiştirir.
name="Mustafa"
name.replace("a","o")
print(name)
```

```python
#split(): bir karakter dizisini verilen parametreye göre böler.
name="Mustafa Cecen"
name.split()
print(name)
name.split(" ",1)
print(name)
```

```python
#lower(): bir karakter dizisindeki tüm karakterleri küçük harfe çevirir.
name="Mustafa"
name.lower()
print(name)
```

```python
#upper(): bir karakter dizisindeki tüm karakterleri büyük harfe çevirir.
name="Mustafa"
name.upper()
print(name)
```

```python
#endswith(): bir karakter dizisinin belirli bir karakterle bitip bitmediğini kontrol eder.
name="Mustafa"
name.endswith("a")
```

```python
#startswith(): bir karakter dizisinin belirli bir karakterle başlayıp başlamadığını kontrol eder.
name="Mustafa"
name.startswith("M")
```

```python
#title(): bir karakter dizisindeki her kelimenin ilk karakterini büyük harfe çevirir.
name="mustafa cecen"
name.title()
```

```python
#join(): bir karakter dizisi listesini belirli bir karakterle birleştirir.
names=["Mustafa","Cecen"]
"-".join(names)
```

```python
#find(): bir karakter dizisinde belirli bir karakterin ilk geçtiği indexi verir.
name="Mustafa"
name.find("a")
```
