```python
#KÜMELER
#Kümeler, sırasız, değiştirilemez ve benzersiz elemanlardan oluşan veri yapılarıdır. 
# Kümeler, süslü parantez {} içinde tanımlanır ve elemanlar virgülle ayrılır. 
# Kümeler, matematikteki kümelere benzer şekilde çalışır ve birçok yerleşik yöntemle birlikte gelir.
#unique elements, unordered, unchangeable
```

```python
mySet = {10,20,30,10,20,40,10,20,40}
print(mySet) 
```

```python
mySet = {10,20,30,10,20,40,10,20,40}
mySet.add(70)
print(mySet) 
```

```python
mySet1 = {10,20,30,40,50}
mySet1 = {30,40,50,60,70}
mySet2 = {50,60,70,80,90}

print(mySet1.union(mySet2)) #birleştirme
print(mySet1.intersection(mySet2)) #kesişim
print(mySet1.difference(mySet2)) #fark
print(mySet1.symmetric_difference(mySet2)) #farkların birleşimi

```

```python
countryList = ["de","fr","tr","fr","tr","tr","de","nl","de","tr"]

print(len(countryList))
print(len(set(countryList)))
```
