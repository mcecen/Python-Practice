```python
boy = int(input("boyunuz nedir(cm)?"))
if boy < 170:
    print("boyunuz kısa")
elif boy < 180:
    print("boyunuz normal")
else:
    print("boyunuz uzun")
```

```python
my_superhero = input("enter superhero: ")

if my_superhero == "Superman":
    print("superman")
elif my_superhero == "Batman":
    print("batman")
elif my_superhero == "Aquaman":
    print("aquaman")
elif my_superhero == "Ironman":
    print("ironman")
else:
    print(":(")
```

```python
my_age=100

if my_age <= 18:
    print("age <= 18")
elif my_age > 18 and my_age <= 30:
    print("18 - age - 30")
elif my_age > 30 and my_age <= 40:
    print("30 - age - 40")
else:
    print("age > 40")
```

```python
my_superhero = input("enter superhero: ")
my_superhero_list = ["Spider Man","Iron Man","Thor","Black Widow"]

if my_superhero in my_superhero_list:
    print(":)")
else:
    print(":(")
```

```python
a = 10
b = 15
c = 20

if a > b or b < c:
    print("superman")
elif a < b and b > c:
    print("batman")
else:
    print("aquaman")
```

```python

my_dictionary = {"k1": 100, "k2":200, "k3":300}

if 200 in my_dictionary.values():
    print("true")
else:
    print("false")
```
