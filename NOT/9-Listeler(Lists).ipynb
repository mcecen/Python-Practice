```python
#LİSTELER
#içerisinde birden fazla veri tutabilen yapılardır. Listeler köşeli parantez [] içerisinde tanımlanır ve elemanlar virgülle ayrılır. Listeler sıralıdır, değiştirilebilir ve tekrar eden elemanlara izin verir.
#farklı türde veriler içerebilirler, örneğin sayılar, metinler, diğer listeler vb. Listeler, Python'da veri yapıları arasında en yaygın kullanılanlardan biridir ve birçok yerleşik yöntemle birlikte gelir.
```

```python
myList = [10,20,30]
print(myList) 
```

```python
x = 10
y = 20
z = 30

myList = [x,y,z]
print(myList)
```

```python
myList = [10,20,30,40,50,60,70]
myList[0] = 100 #mutable, mutability
print(myList) 
```

```python
myList = [10,20,30,40,50,60,70]
print(myList[len(myList)-1]) 
print(myList[-1])
print(myList[-2])
```

```python
myList = [10,20,30,40,50,60,70]
myList.append(80) #listeye eleman ekleme
print(myList)
```

```python
myList = [10,20,30,40,50,60,70]
myList.count(80) #listedeki 80 sayısının kaç tane olduğunu gösterir
```

```python
myList = [10,20,30,40,50,60,70]
myList.insert(2,35) #insert(index, value) index numarasına value değerini ekler
print(myList)
```

```python
myList = [10,20,30,40,50,60,70]
myList.remove(40) #remove(value) listeden value değerini siler
print(myList)
```

```python
myList = [10,20,30,40,50,60,70]
myList.reverse() #listeyi ters çevirir
print(myList)
```

```python
myList = [10,20,30,40,50,60,70]
myList.sort() #listeyi küçükten büyüğe sıralar
print(myList)
```

```python
myNestedList = [10,20,3.14,"mustafa",[1,2,3]]

print(myNestedList[0])
print(myNestedList[1])
print(myNestedList[2])
print(myNestedList[3])
print(myNestedList[4])
print(myNestedList[4][2])
```

```python
firstList = [10,20,30,40,50,60,70]

print(firstList[::])
print(firstList[2::])
print(firstList[:5:])
print(firstList[::2])
print(firstList[1:6:2])
```

```python
list1 = [10,20,30]
list2 = [40,50,60]
combinedList = list1 + list2
print(combinedList)
print(list1*2)
```
